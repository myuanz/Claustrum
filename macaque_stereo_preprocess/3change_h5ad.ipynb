```r
library(Seurat)
library(SeuratDisk)
library(parallel)
library(stringr)

all_file=list.files('/home/liuyx/liuyuxuan/spa_seurat/cla/macaca/renew_spatial_data_240129_macaque/cell_bin_boundary_179_sct')
all_file=str_extract(all_file,'T[0-9]*')
all_file=all_file[!is.na(all_file)]

all_file
length(all_file)
```

```r
workDir='/home/liuyx/liuyuxuan/spa_seurat/cla/macaca/renew_spatial_data_240129_macaque/cell_bin_boundary_179/'
alls=list.files(paste0(workDir))

slice=str_extract(alls,'T[0-9]*')
slice=slice[!(is.na(slice))]
done_file=list.files('/home/liuyx/liuyuxuan/spa_seurat/cla/macaca/renew_spatial_data_240129_macaque/cell_bin_boundary_179_sct_h5ad')

done_slice=done_file[grep('h5ad$',done_file)]
done_slice=str_extract(done_slice,'T[0-9]*')


undone_slice=setdiff(slice,done_slice)
undone_slice
```

```r
for(slice in undone_slice[2:length(undone_slice)]){th_test=readRDS(paste0('/home/liuyx/liuyuxuan/spa_seurat/cla/macaca/renew_spatial_data_240129_macaque/cell_bin_boundary_179_sct/',slice,'_cla_sct.rds'))
snRnaHpSeuratSimplify = th_test
snRnaHpSeuratSimplify@assays = th_test@assays[c("SCT")]
snRnaHpSeuratSimplify@assays$SCT@meta.features$name <- rownames(snRnaHpSeuratSimplify@assays$SCT@meta.features)
DefaultAssay(snRnaHpSeuratSimplify) = "SCT"
SaveH5Seurat(snRnaHpSeuratSimplify, filename = paste0('/home/liuyx/liuyuxuan/spa_seurat/cla/macaca/renew_spatial_data_240129_macaque/cell_bin_boundary_179_sct_h5ad/',slice,"_cla_sct.h5Seurat"))

Convert(paste0('/home/liuyx/liuyuxuan/spa_seurat/cla/macaca/renew_spatial_data_240129_macaque/cell_bin_boundary_179_sct_h5ad/',slice, '_cla_sct.h5Seurat'), dest = "h5ad", overwrite = TRUE)
snRnaHpSeuratSimplify <- NULL
}

```
